## Memuat data

### Subtask:
Muat data dari file Excel ke dalam DataFrame pandas.


**Reasoning**:
Import the pandas library and load the Excel file into a DataFrame.



```python
import pandas as pd

df = pd.read_excel('/content/Online Retail.xlsx')
```

Display the first few rows of the loaded DataFrame to verify the data is loaded correctly.

## Mengeksplorasi data

### Subtask:
Tampilkan beberapa baris pertama data dan periksa informasi umum seperti tipe data, nilai yang hilang, dan statistik deskriptif.


**Reasoning**:
Display the first 5 rows of the dataframe, check the data types and non-null values, and display descriptive statistics for numerical columns.



```python
display(df.head())
display(df.info())
display(df.describe())
```

## Membersihkan data

### Subtask:
Tangani nilai-nilai yang hilang dan hapus duplikat (jika ada).


**Reasoning**:
Check for missing values, fill missing values in 'Description', drop rows with missing 'CustomerID', check for duplicates, drop duplicates, and display info.



```python
print("Missing values before handling:")
display(df.isnull().sum())

df['Description'].fillna('Unknown', inplace=True)
df.dropna(subset=['CustomerID'], inplace=True)

print("\nMissing values after handling:")
display(df.isnull().sum())

print("\nNumber of duplicate rows before handling:")
display(df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("\nNumber of duplicate rows after handling:")
display(df.duplicated().sum())

print("\nDataFrame info after handling missing values and duplicates:")
df.info()
```

## Mengubah tipe data

### Subtask:
Konversi kolom-kolom dengan tipe data yang tidak sesuai (misalnya, mengonversi kolom tanggal ke tipe data datetime).


**Reasoning**:
Convert the `InvoiceDate` column to datetime objects and verify the data type.



```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
print("Data type of InvoiceDate after conversion:")
display(df['InvoiceDate'].dtype)
```

## Menangani outlier

### Subtask:
Identifikasi dan tangani outlier (jika perlu).


**Reasoning**:
Identify numerical columns and calculate descriptive statistics to understand the distribution and potential outliers.



```python
numerical_cols = ['Quantity', 'UnitPrice']
print("Descriptive statistics for numerical columns:")
display(df[numerical_cols].describe())
```

**Reasoning**:
Address negative and zero values in 'Quantity' and 'UnitPrice' as they represent non-standard transactions or errors, then handle extreme outliers using quantile-based filtering.



```python
# Filter out rows with negative or zero Quantity and UnitPrice
df_filtered = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)].copy()

# Calculate quantiles for outlier handling
q_quantity_high = df_filtered['Quantity'].quantile(0.99)
q_unitprice_high = df_filtered['UnitPrice'].quantile(0.99)

# Filter out extreme outliers based on quantiles
df_cleaned = df_filtered[(df_filtered['Quantity'] <= q_quantity_high) & (df_filtered['UnitPrice'] <= q_unitprice_high)].copy()

print("Descriptive statistics after handling outliers:")
display(df_cleaned[numerical_cols].describe())
```
